# Load Datasets from Hugging Face

```python
from datasets import load_dataset
from utils import split_data
```

## SQaLe 

```python
# No val/test splits available
sqale = load_dataset("trl-lab/SQaLe-text-to-SQL-dataset")['train']
sqale
```

### Create Splits

```python
train_frac = 0.8
val_frac = 0.1
train, val, test = split_data(sqale, train_frac, val_frac, random_seed=42)
print(f"Train size: {len(train)}")
print(f"Validation size: {len(val)}")
print(f"Test size: {len(test)}")
```

```python
wikisql = load_dataset("wikisql")
# Includes train, validation, and test splits
wiki_train, wiki_val, wiki_test = wikisql['train'], wikisql['validation'], wikisql['test']
print(f"WikiSQL Train size: {wiki_train.num_rows}")
print(f"WikiSQL Validation size: {wiki_val.num_rows}")
print(f"WikiSQL Test size: {wiki_test.num_rows}")
```
